Un couple valeur propre/vecteur propre est défini de cette manière : $AX = λX$.
Où $A$ est la matrice étudiée, $X$ est le vecteur propre et $λ$ la valeur propre associée à ce vecteur. 
En appliquant cette définition à la matrice $Q$, on obtient $QX = λX$. 

Dans l'énoncé, on définit le vecteur de score par l'équation suivante $r = Qr$. 
En remplaçant $X$ par $r$ dans la définition on obtient $Qr = λr$, ainsi le but est de trouver le vecteur propre $r$ pour une valeur propre λ de 1. 

L'algorithme de la puissance itérée permet donc de calculer une approximation d'une valeur propre ainsi que son vecteur propre associé. 
De plus, lorsque l'algorithme de la puissance itérée est appliqué à une matrice stochastique, il renvoie automatiquement une valeur approchée du vecteur propre de la matrice quand $λ = 1$. 
Ainsi, il faut démontrer que $Q$ est une matrice stochastique.

Une matrice est stochastique si toutes les sommes des valeurs de chaque colonne est égale à 1 et que toutes ses valeurs sont comprises entre 0 et 1.

Démontrons d'abord que les sommes des valeurs de chacunes de ses colonnes sont égales à 1.

$$

Q_j = \displaystyle\sum_{j=1}^N \frac{c_{i,j}}{N_j}  

\equiv Q_j = \displaystyle\frac{1}{N_j}\sum_{j=1}^N c_{i,j}  

$$
De plus, 
$$
N_j = \displaystyle\sum_{j=1}^N c_{i,j}
$$

Ainsi on a :

$$

Q_j = \displaystyle\frac{\displaystyle\sum_{j=1}^N c_{i,j}}{\displaystyle\sum_{j=1}^N c_{i,j}}

\equiv Q_j = 1 

$$

D'où chaque colonne de Q, est égale à 1.

Finalement, nous allons démontrer que toute valeur de Q est comprise entre 0 et 1.

On a donc :

$$q_{i,j} = \displaystyle\frac{c_{i,j}}{N_j}$$ avec $$c_{i,j}\in\ {0,1} $$ et $$ N_j\in\ N$$

Or la division de 1 ou 0 par un nombre entier est forcément comprise entre 0 et 1.

Donc Q remplie les 2 conditions pour être une matrice stochastique.

C'est pourquoi en appliquant l'algorithme de la puissance itérée sur la matrice $Q$, on obtient le vecteur $r$ contenant tous les scores des pages.
